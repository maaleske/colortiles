```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
```

```python
import holoviews as hv
#hv.extension('bokeh')
hv.extension('matplotlib')
```

```python
def slicer(ds, valuedim, dynamic=True):
    hv_ds = hv.Dataset(ds,
                   #kdims=['time', 'band', 'y', 'x'],
                   vdims=[valuedim])
    dim = hv.Dimension(valuedim, soft_range=(ds[valuedim].min(), ds[valuedim].max()))
    # left bottom right top
    extents = (ds.x.min(), ds.y.min(), ds.x.max(), ds.y.max())
    imb = hv_ds.to(hv.Image, ['x', 'y'], dim, extents=extents, dynamic=dynamic)
    imy = hv_ds.to(hv.Image, ['y', 'band'], dim, dynamic=dynamic)
    imx = hv_ds.to(hv.Image, ['x', 'band'], dim, dynamic=dynamic)
    spectra = hv_ds.to(hv.Curve, 'band', dim, dynamic=dynamic)

    layout = imb + spectra# + imy + imx
    return layout.cols(2)
```

```python
%%opts Image [aspect='equal' colorbar=True] (cmap='viridis')
%%output max_frames=10000
slicer(dn_ds.set_coords(['wavelength', 'material']), 'dark_corrected_dn')
```
